```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection

```python
parkinson_data = pd.read_csv('/content/Project 14 Parkinsons Disease.csv')
```

0 --> Healthy

1 --> Parkinson

```python
parkinson_data.drop(columns='name',axis=1,inplace=True)
```

```python
#grouping the data based on status
parkinson_data.groupby('status').mean()
```

Data Preprocessing

```python
X = parkinson_data.drop(columns='status',axis=1)
Y = parkinson_data['status']
```

```python
print(X)
print(Y)
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

Model Training

```python
# SVM
model = svm.SVC(kernel='linear')
```

```python
#training the SVM
model.fit(X_train,Y_train)
```

Model Evaluation

```python
#accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train ,X_train_prediction )
print('Accuracy Score of training data is',training_data_accuracy)
```

```python
#accuracy score on training data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test ,X_test_prediction)
print('Accuracy Score of testing data is',testing_data_accuracy)
```

Buliding a Predictive System

```python
input_data = (148.09000,162.82400,67.34300,0.00762,0.00005,0.00467,0.00354,0.01400,0.05428,0.49700,0.03357,0.02494,0.03635,0.10070,0.02431,21.71800,0.487407,0.727313,-6.261141,0.120956,2.137075,0.141958)

#changing input to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshaping the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

#standardize data
std_data = scaler.transform(input_data_reshaped)

prediction = model.predict(std_data)

print(prediction)

if(prediction[0]==0):
  print('The Person is Healthy')
else:
  print('The Person has Parkinson')
```
